# What is Softmax?

Softmax is a transformation, when applied to range of logits (the outputs of a neural network layer), turns their real values into probabilities that:

-  together sum up to one
- each sit between $0\ldots 1$

Often, for classification tasks, it will be used as the final output layer of a neural network.

Imagine outputs $x_{1}$, $x_{2}$, ...

The softmax for one of these numbers $x_{i}$ is:
$$s(x_{i}) = \frac{e^{x_{i}}}{ \sum_{j=1}^{n} e^{x_{j} }}$$

Look at the function $e^{x}$ plotted out:

```python
#| echo: false
import numpy as np

import matplotlib.pyplot as plt

x = np.linspace(-3, 3, 100)
y = np.exp(x)

plt.plot(x, y)
plt.title(r'Plot of $e^{x}$')
plt.xlabel('x')
plt.ylabel(r'$e^{x}$')
plt.grid(True)
plt.show()
```

Oberve that this function maps x from $-\infty$ to $+\infty$ to a range between $0\ldots1$.  It also "encourages" large values proportionally more than small values.

# Relation to Cross-Entropy Loss

Recall the formula for Cross-Entropy Loss, as outlined in [Cross-Entropy Loss](./cross-entropy-loss.ipynb):

$$\mathcal{L} = -\sum_{c=1}^{M} y_c \log(p_c)$$
where:

- $M$ is the number of categories
- $y_c$ is the ground truth (one-hot)
- $p_c$ is the predicated probability for class c (after applying softmax)

$$\mathcal{L} = -[ y_1 log(p_1) + y_2 log(p_2) + ... + y_M log(p_M) ] $$

however, for only one single category the ground truth $y_{\text{true}}$ will be $1$ for all the others, the ground truth $y_i$ will be $0$.  This means we can further simplify this to:

$$\mathcal{L} =  -[ y_{\text{true}} log(p_\text{true})] $$
